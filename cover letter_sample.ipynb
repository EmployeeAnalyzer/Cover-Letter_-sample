## 의미 연결망 분석(SNA)

```python
#인접행렬 생성
import numpy as np
adj_matrix = np.zeros((num_top_nouns,num_top_nouns), int)
for sentence in sen_nouns:
    for wi, i in word2id.items():
        if wi in sentence:
            for wj, j in word2id.items():
                if i != j and wj in sentence:
                    adj_matrix[i][j] += 1
adj_matrix
```

```python
network = nx.from_numpy_matrix(adj_matrix)
list(network.adjacency())
```

```python
from matplotlib import font_manager as fm
from matplotlib import rc

# font_path = r'C:\Users\이선주\OneDrive - 한양대학교\바탕 화면\WISET\NanumSquareRoundR.ttf'
# font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family ='Malgun Gothic')

fig = plt.figure()
fig.set_size_inches(20,20)
ax = fig.add_subplot(1,1,1)
ax.axis('off')
option={
    'node_color' : 'yellow',
    'node_size' : 2000,
#      'size' : 2
}
nx.draw(network, labels= id2word, font_family ='Malgun Gothic' , ax=ax, **option)
```

```python
#Degree
nx.degree_centrality(network)
```
